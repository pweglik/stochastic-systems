```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data = pd.read_pickle("/Users/woj/agh-stochastic-systems/CEM-RL/results/HalfCheetah-v4-run1/log.pkl")
```

```python



# Setting the style with seaborn
sns.set(style="whitegrid")


# Create a plot with seaborn and make background transparent
fig, ax = plt.subplots(figsize=(10, 6))
ax.plot(data["total_steps"], data["average_score"], marker='o', linestyle='-', label='Average Score')
# ax.plot(data["total_steps"], data["average_score_rl"], marker='s', linestyle='--', label='Average Score RL')
# ax.plot(data["total_steps"], data["average_score_ea"], marker='^', linestyle='-.', label='Average Score EA')
ax.plot(data["total_steps"], data["best_score"], marker='*', linestyle=':', label='Best Score')

# Adding titles and labels
ax.set_title('Performance Metrics Over Training Steps')
ax.set_xlabel('Total Training Steps')
ax.set_ylabel('Fitness')
ax.legend()

# Save as SVG with transparent background
plt.tight_layout()
fig.patch.set_alpha(0) # Making the background of the figure transparent
ax.patch.set_alpha(0) # Making the background of the axes transparent
plt.savefig("asdasdsad.svg", format='svg', transparent=True)

```
